```python
import json
import numpy as np
from pandas import DataFrame
```

```python
def loadResults(files):
    Results = []
    for ResultFile in files:
        with open(ResultFile) as F:
            for line in F:
                jsonObj = json.loads(line)
                Results.append(jsonObj)
    return Results

def collapse(arr):
    return arr[0] + "." + arr[1]

def flatten(d, prefix=''):
    result = {}
    for k, v in d.items():
        if isinstance(v, dict):
            result.update(flatten(v, prefix=k+"."+prefix))
        else:
            result[prefix+k] = v
    return result

Results = loadResults(["result.st_test_native.txt", "result.sa_pysuffix.txt", "result.st_mccreight.txt", "result.fm_index_plusplus.txt",
                       "result.sa_linsuffarr.txt", "result.fm_index.txt", "result.sa_simple_accel.txt", "result.sa_sais.txt", "result.sa_index_accel.txt"])
```

```python
Languages = ["genomic", "english", "chinese"]
TestClasses = []
for a in [x["info"]["test_class"] for x in Results]:
    if a not in TestClasses:
        TestClasses.append(a)
TestClasses = sorted(TestClasses)
print "Langs:", Languages
print "TestClasses:", TestClasses
```

```python
def tableify(Results):
    tables = []
    for Lang in Languages:
        table = []
        for TestClass in TestClasses:
            SubResults = [x for x in Results if (Lang in x['info']['corpus_filepath']) and (x['info']['test_class'] == TestClass)]
            TextLens = sorted(set([x["text_len"] for x in SubResults]))
            SearchLens = sorted(set([x["avg_search_len"] for x in SubResults]))
            for TextLen in TextLens:
                for SearchLen in SearchLens:
                    Res = [x for x in SubResults if (x["text_len"] == TextLen) and (x["avg_search_len"] == SearchLen)]
                    if len(Res) > 0:
                        Res = Res[0]
                        row = [collapse(TestClass), TextLen, SearchLen, 
                               Res["size"],
                               Res["build_time"]["KernelTime"] + Res["build_time"]["UserTime"],
                               Res["search_time"]["KernelTime"] + Res["search_time"]["UserTime"],
                               Res["info"]["num_searches"] * Res["info"]["lookups_per_query"]]
                        table.append(row)
        tables.append(table)
    return tables

tables = tableify(Results)
col_names = ["Implementation", "Text Length", "Search Length", "Structure Size (Bytes)", "Build Time", "Search Time", "Total Queries"]
table_df1 = DataFrame(tables[0], columns=col_names)
table_df2 = DataFrame(tables[1], columns=col_names)
table_df3 = DataFrame(tables[2], columns=col_names)

with open('results.html', 'w') as result_html:
    result_html.write("<style type=\"text/css\">table {font-family: verdana,arial,sans-serif;font-size:11px;color:#333333;border-width: 1px;	border-color: #666666;	border-collapse: collapse;}table th {border-width: 1px;padding: 8px;border-style: solid;border-color: #666666;background-color: #dedede;}table.gridtable td {border-width: 1px;padding: 8px;border-style: solid;border-color: #666666;background-color: #ffffff;}</style>")
    result_html.write("<h1>Genomic</h1>")
    result_html.write(table_df1.to_html())
    result_html.write("<h1>English</h1>")
    result_html.write(table_df2.to_html())
    result_html.write("<h1>Chinese</h1>")
    result_html.write(table_df3.to_html())
```

```python
#x axis is text len, y is time to build, lines are implementations
GraphColors = {
    #Trees red
    "st_mccreight.STMcCreightTest": ['r', '-'],
    "st_test_native.SuffixTreeTest": ['r', '--'],

    #Arrays blue
    "sa_linsuffarr.SuffixArrayTest": ['b', '-'],
    "sa_pysuffix.SuffixArrayTest": ['b', '--'],
    
    "sa_sais.SuffixArraySAIS": ['c', '-'],
    "sa_variants.simple_accel.SuffixArrayTest": ['c', '--'],
    "sa_variants.index_accelerant.SuffixArrayTest": ['m', '-'],

    #FM Index green
    "fm_index.FMIndexTest": ['g', '-'],
    "fmindexplusplus.src.fmindex_test.FMIndexTest": ['g', '--']
}
def drawBuildTimeGraph(Results):
    for Lang in Languages:
        plt.figure(figsize=[8,6])
        for TestClass in TestClasses:
            SubResults = [x for x in Results if (Lang in x['info']['corpus_filepath']) & (x['info']['test_class'] == TestClass)]
            BuildTimes = [[r['text_len'], r['build_time']['KernelTime'] + r['build_time']['UserTime']] for r in SubResults]
            BuildTimesC = [[a, [x[1] for x in BuildTimes if x[0] == a]] for a in [i for i in sorted(set([x[0] for x in BuildTimes]))]]
            xs = [x[0] for x in BuildTimesC]
            ys = [np.average(x[1]) for x in BuildTimesC]
            yerrMin = [np.average(x[1]) - np.min(x[1]) for x in BuildTimesC]
            yerrMax = [np.max(x[1]) - np.average(x[1]) for x in BuildTimesC]
            if len(xs) > 0:
                plt.errorbar(xs, ys, marker='.', yerr=[yerrMin, yerrMax], label=collapse(TestClass), color=GraphColors[collapse(TestClass)][0], ls=GraphColors[collapse(TestClass)][1])
        plt.title(Lang.title() + " - Build Time")
        plt.xscale('log')
        plt.yscale('log')
        plt.axes().xaxis.grid()
        plt.axes().yaxis.grid()
        plt.xlim(xmin=48000)
        plt.ylim(ymin=0.002)
        plt.xlabel('Length of Indexed String (characters)')
        plt.ylabel('Build Time (seconds)')
        plt.legend(bbox_to_anchor=(1, 1), loc=2)
        plt.show()
drawBuildTimeGraph(Results)
```

```python
def drawSpaceUsageGraph(Results):
    for Lang in Languages:
        plt.figure(figsize=[8,6])
        for TestClass in TestClasses:
            if TestClass == ['fmindexplusplus.src.fmindex_test', 'FMIndexTest']:
                continue
            SubResults = [x for x in Results if (Lang in x['info']['corpus_filepath']) & (x['info']['test_class'] == TestClass)]
            BuildTimes = [[r['text_len'], r['size']] for r in SubResults]
            BuildTimesC = [[a, [x[1] for x in BuildTimes if x[0] == a]] for a in [i for i in sorted(set([x[0] for x in BuildTimes]))]]
            xs = [x[0] for x in BuildTimesC]
            ys = [np.average(x[1]) for x in BuildTimesC]
            yerrMin = [np.average(x[1]) - np.min(x[1]) for x in BuildTimesC]
            yerrMax = [np.max(x[1]) - np.average(x[1]) for x in BuildTimesC]
            if len(xs) > 0:
                plt.errorbar(xs, ys, marker='.',  yerr=[yerrMin, yerrMax], label=collapse(TestClass), color=GraphColors[collapse(TestClass)][0], ls=GraphColors[collapse(TestClass)][1])
        plt.title(Lang.title() + " - Memory Usage")
        #plt.plot([64000, 64000000], [64000, 64000000], color='k', label="Corpus Size")
        
        plt.xscale('log')
        plt.yscale('log')
        plt.axes().xaxis.grid()
        plt.axes().yaxis.grid()
        plt.xlim(xmin=48000)
        plt.ylim(ymin=0)
        plt.xlabel('Length of Indexed String (characters)')
        plt.ylabel('Memory Usage (bytes)')
        plt.legend(bbox_to_anchor=(1, 1), loc=2)
        plt.show()
drawSpaceUsageGraph(Results)
```
